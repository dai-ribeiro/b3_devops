Q06 - Um estacionamento cobra um valor mínimo de 10 reais, correspondente a 1h de uso. Cada hora adicional gera mais 5 reais de cobrança. Ex: um carro estacionado por 5 horas irá pagar 10 reais pela primeira hora + 5 reais pela segunda + 5 pela terceira + 5 pela quarta + 5 pela quinta, totalizando 30 reais.

Faça um programa que pergunte para a usuária quanto tempo seu carro ficou estacionado e responde o valor em reais a ser pago.

OBS: o estacionamento não considera minutos, tampouco horas fracionárias. Portanto, o seu programa não precisa se preocupar com isso. Se a pessoa ficou 5 horas e 2 minutos, ela passou de 5 horas, portanto ela deve digitar a hora inteira mais próxima: 6 horas.

```python
tempo = int(input('Informe quantas horas ficou estacionado: '))

hora_1 = 10
hora_adicional = 5
valor_adicional = (tempo - 1) * hora_adicional

valor = hora_1 + valor_adicional

print(f'Valor a pagar: R$ {valor}')
```

Q07 - Faça um programa que pergunta a temperatura em graus Celsius e responde a temperatura correspondente em graus Fahrenheit.

Q09 - Vai um cupom de desconto aí?

Q10 - Faça um programa que pergunta o salário da funcionária, o aumento a ser aplicado (em %) e responde o seu novo salário.

```python
salario = float(input('Informe o valor do salário: '))
aumento = float(input('Informe o percentual de aumento: '))

percentual = aumento / 100

novo_salario = salario * (1 + percentual)

print(f'O novo salário será: R$ {novo_salario}')
```

Q11 - Vamos fazer um programa para calcular o rendimento de uma aplicação.

Faça um programa que pergunta quanto dinheiro a usuária irá aplicar, a taxa de juros ao mês (em %) e a duração da aplicação (em meses).

Seu programa deve responder as seguintes informações:

Qual o valor total a ser sacado pela usuária ao final da aplicação?
Quantos reais a pessoa recebeu apenas de juros?
Quantos % a aplicação rendeu no total?
Atenção: ao buscar as fórmulas para utilizar no problema, busque pela fórmula de juros compostos. Não utilize a fórmula de juros simples.

```python
capital = float(input('Informe o valor a ser aplicado: '))
taxa = float(input('Informe a taxa de juros mensal: ')) / 100
tempo = int(input('Informe o tempo da aplicação em meses: '))

montante = capital * ((1 + taxa)**tempo)
juros = montante - capital
rendimento = (juros / capital) * 100

print(f'O valor final da aplicação será R$ {montante:.2f}')
print(f'O valor dos juros será R$ {juros:.2f}')
print(f'A rendimento da aplicação é de {rendimento:.2f}%')
```
